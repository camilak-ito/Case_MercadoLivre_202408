### Explode a coluna 'attributes' em um novo dataframe e faz *left join*

```python
# Cria um novo DF a partir do original
df_attributes = df_original[["id", "attributes"]]

# Converte a coluna em um array
df_attributes["attributes"] = df_attributes["attributes"].apply(literal_eval)

# Explode o array e extrai os valores que nos interessam
df_attributes = df_attributes.explode("attributes")
df_attributes['attr_id'] = df_attributes["attributes"].apply(lambda x: x.get('id'))
df_attributes['attr_value_name'] = df_attributes["attributes"].apply(lambda x: x.get('value_name'))

# Dropa a coluna original
df_attributes = df_attributes.drop(["attributes"], axis=1)
# df_attributes.head()

# Pivota o df
df_attributes = df_attributes.pivot(index='id', columns='attr_id', values='attr_value_name')
# df_attributes.head()

df_merged = df_original.merge(df_attributes, how="left", on="id")
df_merged.shape
```

### Analises

```python
# Quantidade de produtos por marca
df_merged.groupby("BRAND").describe()
```

```python
# Quantidade de produtos por marca e memoria
df_merged[["BRAND","INTERNAL_MEMORY","price"]].groupby(["BRAND","INTERNAL_MEMORY"]).agg(['min', 'max','mean','count'])
```

```python
# Quantidade de produtos por preco, menor, maior e a média
df_merged[["BRAND","LINE","price"]].groupby(["BRAND","LINE"]).agg(['min', 'max','mean','count'])
```

```python
# Quantidade de produtos por condição (Novo, Usado)
df_merged[["BRAND","ITEM_CONDITION","id"]].groupby(["BRAND","ITEM_CONDITION"]).agg(['count'])
```
